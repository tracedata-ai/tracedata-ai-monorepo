```python
from langgraph.graph import StateGraph, END
from typing import TypedDict
```

```python
# 1. Define your state (what data flows through the graph)
class MyState(TypedDict):
    message: str 
```

```python
# 2. Define nodes (functions that process the state)
def hello_node(state: MyState):
    return {"message": state["message"] + " Hello"}

def world_node(state: MyState):
    return {"message": state["message"] + " World!"}
```

```python
# 3. Build the graph
graph = StateGraph(MyState)
graph.add_node("hello", hello_node)
graph.add_node("world", world_node)
```

```python
# 4. Define the flow
graph.set_entry_point("hello")
graph.add_edge("hello", "world")
graph.add_edge("world", END)
```

```python
# 5. Compile and run
app = graph.compile()
result = app.invoke({"message": "Starting:"})
print(result["message"])  # Output: "Starting: Hello World!"
```
