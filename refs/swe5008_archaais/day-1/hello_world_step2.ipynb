```python
from langgraph.graph import StateGraph, END
from typing import TypedDict
```

```python
# Enhanced state with multiple fields
class ProcessingState(TypedDict):
    message: str
    step_count: int
    history: list[str]
```

```python
# Nodes that do more interesting work
def process_step1(state: ProcessingState):
    new_message = state["message"].upper()
    return {
        "message": new_message,
        "step_count": state["step_count"] + 1,
        "history": state["history"] + ["Step 1: converted to uppercase"],
    }


def process_step2(state: ProcessingState):
    new_message = f"[{state['message']}], brackerts are added!"
    return {
        "message": new_message,
        "step_count": state["step_count"] + 1,
        "history": state["history"] + ["Step 2: added brackets"],
    }
```

```python
# Build and run
graph = StateGraph(ProcessingState)
graph.add_node("step1", process_step1)
graph.add_node("step2", process_step2)
graph.set_entry_point("step1")
graph.add_edge("step1", "step2")
graph.add_edge("step2", END)
```

```python
app = graph.compile()
result = app.invoke({"message": "hello world", "step_count": 0, "history": []})
print(result)
```
